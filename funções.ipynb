```
isadora = ["isa", "dora", 21, 1.999, True]
```

```
def print_index_value_type(index, value):
    '''
        Function for printing information contained in list iteration loops.

        Parameters:
        index: index (key) from list
        value: values in the list
    '''
    print('_______________________')
    print(f'Index - {index}')
    print(f'Value - {value}')
    print(f'Type - {type(value)}')


help(print_index_value_type)
```

```
# Aqui, obtivemos index e value através de uma iteração, feita pelo for, na lista isadora, e pela função enumerate.

for index_value in enumerate(isadora):
    print_index_value_type(index_value[0], index_value[1])
```

```
# Aqui, obtivemos index e value através da iteração da lista isadora, feita pelo for, e pela função enumerate.

for index, value in enumerate(isadora):
    print_index_value_type(index, value)  
```

```
# Aqui, obtivemos o index através de uma iteração feita pelo for, a qual começa em 0 e finaliza com o número que representa o tamanho da lista isadora.

for index in range(len(isadora)):
    print_index_value_type(index, isadora[index])
```

```
# Aqui, obtivemos index e value através de uma iteração feita pelo while, e pelo contador count, que foi incrementado.

count = 0
while count < len(isadora):
    print_index_value_type(count, isadora[count])
    count += 1 
```

```
# Aqui, obtivemos o value através da iteração da lista isadora, feita pelo for. E o index foi obtido através do contador count, que foi incrementado.

count = 0
for value in isadora:
    print_index_value_type(count, value)
    count += 1    
```
